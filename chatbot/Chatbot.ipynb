##Step 2:

```python
dataset = data['code']
```

```python
data=data.applymap(lambda x:x.lower() if type(x)==str else x)
```

```python
data=data.applymap(lambda x:' '.join(x.split()).strip() if type(x)==str else x)
```

```python
data.head()
```

```python
from textblob import TextBlob
data=data.applymap(lambda x:str(TextBlob(x)) if type(x)==str else x)
```

```python
!pip install contractions
import contractions

data=data.map(lambda x:contractions.fix(x) if type(x)==str else x)
```

```python
import re
data=data.map(lambda x:re.sub(r'[^\w\s]','',x) if type(x)==str else x)
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
data=data.map(lambda x:word_tokenize(x) if type(x)==str else x)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
data=data.map(lambda x:[word for word in x if word not in stopwords.words('english')] if type(x)==list else x)
```

```python
data.head()
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer=WordNetLemmatizer()
data=data.map(lambda x:[lemmatizer.lemmatize(word) for word in x] if type(x)==list else x)
```

##Step 3:

```python
data['question']=data['question'].apply(lambda x:' '.join(x) if isinstance(x, list) else str(x))
data['code']=data['code'].apply(lambda x:' '.join(x) if isinstance(x, list) else str(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
vector=vectorizer.fit_transform(data['question'])
```

```python
data['question']
```

##Step 4:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import pandas as pd
```

```python
knn = NearestNeighbors(n_neighbors=3, metric='cosine')
knn.fit(vector)
```

```python
data['original_code']=dataset
```

```python
def get_predicted_code(query):
    query_vector = vectorizer.transform([query])
    distances, indices = knn.kneighbors(query_vector)
    closest_indices = indices[0]
    closest_codes = data.iloc[closest_indices]['original_code'].tolist()
    return closest_codes
```

```python
new_question = "Formulate an equation to calculate the height of a triangle given the angle, side lengths and opposite side length."
predicted_codes = get_predicted_code(new_question)
```

```python
print("Answer: "+ str(predicted_codes[0]))
```
